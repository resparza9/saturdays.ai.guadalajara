```python
# Lee el dataset original
csv_original = 'DataSetNike.csv'
dataset_original = pd.read_csv(csv_original, encoding='latin-1', low_memory=False)
```

```python
# Crea un nuevo dataframe
dataset_deseado2 = dataset_original[['Material', 'Date', 'Units']]
```

```python
# Elimina los registros con unidades negativas
dataset_deseado2 = dataset_deseado2[dataset_deseado2.Units >= 1]
```

```python
# Renombra la columna Matrial por Código
dataset_deseado2 = dataset_deseado2.rename(columns = {'Material' : 'CÓDIGO'})
```

```python
# Convierte el campo Date que es un string a formato Datetime
dataset_deseado2['Date'] = pd.to_datetime(dataset_deseado2['Date'])
```

```python
# Obtener Año y Mes de la fecha
dataset_deseado2['Month'] = dataset_deseado2['Date'].dt.month
dataset_deseado2['AÑO'] = dataset_deseado2['Date'].dt.year
#dataset_deseado2
```

```python
# Se genera un dataframe temporal con la conversion a tabla dinámica
# Se obtiene una variable para la sumatoria de unidades por mes 
temporal = dataset_deseado2.pivot_table(values = 'Units', index = ['CÓDIGO', 'AÑO'],
           columns = 'Month', aggfunc = np.sum)

# Inicializa los valor NA con el valor 0
temporal = temporal.fillna(0)
#temporal
```

```python
# Inicializa el indice del dataframe temporal
temporal = temporal.reset_index()
```

```python
# Renombra los valores numéricos de los meses por un valor en texto
#temporal = temporal.rename(columns =
#{1:'ENE.',2:'FEB.',3:'MAR.',4:'ABR.',5:'MAY.',6:'JUN.',7:'JUL.',8:'AGO.',9:'SEP.',10:'OCT.',11:'NOV.',12:'DIC.'})

# Convertir a int la suma de unidades por mes : Se que hay mejores formas de hacerlo pero de
# momento planteo esta solución directa por la premura, pero no eficiente.
temporal[['ENE.','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.','OCT.','NOV.','DIC.']] = temporal[['ENE.','FEB.','MAR.','ABR.','MAY.','JUN.','JUL.','AGO.','SEP.','OCT.','NOV.','DIC.']].astype(int)
#temporal
```

```python
# Ordena el dataframe temporal y lo ordena por Año y Código para cumplir con la agrupación
dataset_deseado2 = temporal.sort_values(['AÑO','CÓDIGO'])
dataset_deseado2 = dataset_deseado2.reset_index()
#dataset_deseado2["Month"]
```

```python
# Se destruye el dataframe temporal
del(temporal)
```
